$V_{cilindro} = \pi*r^{2}*h$

```python
import numpy as np

def volumen():
  r = float(input('Escriba el radio en cm: '))
  h = float(input('Escriba el altura en cm: '))

  v = np.pi* r**2 * h
  round(v, 2)

  print('El volumen del cilindro es: ' + str(round(v, 2)) + ' centimetros cúbicos')
```

```python
volumen()
```

```python
def volumen(r, h):

  v = np.pi* r**2 * h
  round(v, 2)

  print('El volumen del cilindro es: ' + str(round(v, 2)) + ' centimetros cúbicos')
```

```python
volumen(7.4, 5.2)
```

# Bucles

```python
contador = 0

while contador < 5:
  print(contador)

  contador += 1

print('Fuera del ciclo')
```

```python
# Imprimir las potencias de 2

i = 50

while i >= 0:
  potencia = 2**i
  print('2 elevado a {} es {}'.format(i, potencia))
  i -= 1

print('Terminada la ejecución.')
```

```python
for i in range(20, 51, 5):
  potencia = 2**i
  print('2 elevado a {} es {}'.format(i, potencia))
```
