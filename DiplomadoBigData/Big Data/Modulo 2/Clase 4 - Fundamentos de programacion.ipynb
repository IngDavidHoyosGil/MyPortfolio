# Diccionarios

```python
champions = {
    'Real Madrid' : 13,
    'Milan' : 7,
    'Barcelona' : 5,
    'Liverpool' : 3,
    'Bayern' : 6,
    'Jugadores' : ['Messi', 'Cristiano', 'Neymar', 'Mbape']
}
```

```python
type(champions)
```

```python
champions.keys()
```

```python
champions.values()
```

```python
champions.items()
```

```python
champions['Jugadores'][3]
```

```python
for i in champions.keys():
  print(i)
```

```python
for i in champions.values():
  print(i)
```

```python
for i,j in champions.items():
  print(i + ' -> ' + str(j))
```
