```python
x = int(input())
y = int(input())

if (x > y):
  print('x es mayor a y')
elif (x < y):
  print('x es menor a y')
else:
  print('x es igual a y')
```

```python
nombre = input('¿Cuál es su nombre?: ')
edad = int(input('¿Cuántos años tienes?: '))

if edad > 18:
  print('Hola ' + nombre + ' , tienes ' + str(edad) + ' años de edad. Eres mayor de edad')
elif edad == 18:
  print('Hola ' + nombre + ' , tienes ' + str(edad) + ' años de edad. Haz cumplido la mayoría de edad.')
else:
  print('Hola ' + nombre + ' , tienes ' + str(edad) + ' años de edad. Eres menor de edad')
```
