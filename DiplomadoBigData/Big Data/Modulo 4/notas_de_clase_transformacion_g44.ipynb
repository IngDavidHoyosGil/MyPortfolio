```python
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
cat_ord = ['IsActiveMember','Exited']
cat_nomin = ['Geography','Gender','HasCrCard'] # 'Surname'
num_estand = []
num_norm = []
num_nada = []

numeric_var = ['Age','Balance','EstimatedSalary' ]

X = df[cat_ord + cat_nomin + numeric_var]
```

```python
X.head()
```

```python
X.info()
```

```python
pipeline_categoricas_nom = Pipeline(
    [('imputador', SimpleImputer(strategy = 'most_frequent')),
     ('dummy',OneHotEncoder(handle_unknown='ignore'))]
)

pipeline_categoricas_nom
```

```python
pipeline_categoricas_ord = Pipeline(
    [('imputador', SimpleImputer(strategy = 'most_frequent')),
     ('ordinal',OrdinalEncoder())]
)

pipeline_categoricas_ord
```

```python
# Pipeline para variables numricas
pipeline_numerico = Pipeline(
    [('imputador', SimpleImputer(strategy = 'median')),
     ('escalador', StandardScaler())]
     )

pipeline_numerico
```

```python
procesador = ColumnTransformer(
    [('codificador ordinal',pipeline_categoricas_ord,cat_ord),
     ('codificador nominal',pipeline_categoricas_nom,cat_nomin),
     ('codificador numerico',pipeline_numerico,numeric_var)]
)

procesador
```

```python
X
```

```python
procesador.fit(X)
X_transform = procesador.transform(X)
X_transform.shape
```

```python
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

# Datos categricos
data = np.array([
    ['bajo', 'azul', 'triangulo'],
    ['alto', 'rojo', 'circulo'],
    ['mediano', 'verde', 'circulo'],
    ['alto', 'rojo', 'cuadrado']
])

# Especificar el orden de las categoras para cada caracterstica
categories = [
    ['bajo', 'mediano', 'alto'],         # Orden para la primera caracterstica
    ['azul', 'rojo', 'verde'],           # Orden para la segunda caracterstica
    ['triangulo', 'circulo', 'cuadrado'] # Orden para la tercera caracterstica
]

# Crear el codificador ordinal con las categoras especificadas
encoder = OrdinalEncoder(categories=categories)

# Ajustar el codificador a los datos y transformarlos
encoded_data = encoder.fit_transform(data)

# Mostrar los datos originales y codificados
print("Datos originales:")
print(data)
print("\nDatos codificados:")
print(encoded_data)
```
