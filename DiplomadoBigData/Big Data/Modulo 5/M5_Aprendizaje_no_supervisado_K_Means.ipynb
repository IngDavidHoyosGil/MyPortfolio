## **5. Evaluación del desempeño del agrupamiento**

### **5.1 Coeficiente silueta**

El coeficiente silueta es un indicador de similitud, toma valores entre -1 y 1. Donde 1 es el indicador óptimo de separación. Este coeficiente mide la cohesión entre datos del mismo grupo y la separación entre datos de diferentes grupos.


El coeficiente de silueta combina la distancia media *intra-cluster* ($a$) y la distancia media al grupo más cercano ($b$) para cada muestra ($s_i$):
$$ s_i = \frac{b - a}{\max(a,b)}$$

Valores cercanos a 1 indican que la distancia inter-cluster es mucho más grande que la distancia intra-cluster.

### **5.2 Inercia**

La inercia o distancia intra-clúster mide la cohesión en cada clúster, el algoritmo de k-means busca disminuir esta medida en función de las iteraciones.

$$\sum_{i=0}^{n}\min_{\mu_j \in C}(||x_i - \mu_j||^2)$$

Donde $\mu_j$ es la media de cada clúster, también llamado centroide.

Ahora usemos las métricas para encontrar cuál es la agrupación óptima.

```python
from sklearn.metrics import silhouette_score

coef_silueta = []
inertia = []

K = np.arange(2, 1000)

for numero_clusters in K:
  km = KMeans(n_clusters = numero_clusters,
              random_state= 11,
              max_iter = 1000)

  km.fit(X_blobs)

  etiqueta = km.labels_
  coef_silueta.append(
      silhouette_score(X_blobs, etiqueta)
  )
  inertia.append(km.inertia_)
```

```python
fig, axes = plt.subplots(1,2,figsize = (8,4),constrained_layout=True)

max_vals = 25

fig.suptitle('Evaluación del desempeño dataset Blobs')

axes[0].plot(K , coef_silueta, marker='x', c = 'r')
# axes[0].plot(K , coef_silueta, c = 'r')
axes[0].set_xlabel('Número de Cluster')
axes[0].set_ylabel('Coeficiente silueta')
axes[0].set_xlim((0,max_vals))

axes[1].plot(K , inertia, marker='x')
# axes[1].plot(K , inertia)
axes[1].set_xlabel('Número de Cluster')
axes[1].set_ylabel('Inercia')
axes[1].set_xlim((0,max_vals))
```

Para la escogencia óptima del número de clusters se busca:
* Maximizar el coeficiente silueta
* Disminuir la inercia

A menudo se usa el método del codo en la gráfica de inercia, es decir el punto máximo de inflexión. En el caso del dataset blobs está entre 3 y 4. En 3 y 4 también se obtienen coeficientes silueta relativamente altos indicado una separación eficiente entre clusters y cohesión interna entre cada grupo de datos.

```python
from sklearn.metrics import silhouette_score

coef_silueta = []
inertia = []

K = range(2,14)
for numero_clusters in K:
  km = KMeans(n_clusters = numero_clusters, random_state= 11, max_iter = 1000)
  km.fit(X_iris)
  y = km.predict(X_iris)
  coef_silueta.append(silhouette_score(X_iris,y))
  inertia.append(km.inertia_)
```

```python
fig, axes = plt.subplots(1,2,figsize = (10,4))

fig.suptitle('Evaluación del desempeño dataset Iris')

axes[0].plot(K , coef_silueta, marker='x', c = 'r')
axes[0].set_xlabel('Número de Cluster')
axes[0].set_ylabel('Coeficiente silueta')

axes[1].plot(K , inertia, marker='x')
axes[1].set_xlabel('Número de Cluster')
axes[1].set_ylabel('Inercia')
```

* Con método merge

```python
iris_final_2 = pd.DataFrame(X_iris)
iris_final_2.head(2)
```

```python
iris_etiquetas = pd.DataFrame(km_final.labels_, columns=['etiqueta'])
print(iris_etiquetas.head(2))

iris_etiquetas_og = pd.DataFrame(y_iris, columns=['etiqueta_og'])
print(iris_etiquetas_og.head(2))
```

```python
iris_merged = (iris_final_2
               .merge(iris_etiquetas
                      , how='left'
                      , left_index=True
                      , right_index=True
                      )
               .merge(iris_etiquetas_og
                      , how='left'
                      , left_index=True
                      , right_index=True
                      )
               )

iris_merged['Exactitud'] = iris_merged['etiqueta'] == iris_merged['etiqueta_og']
iris_merged.head(2)
```

```python
pd.crosstab(iris_merged['etiqueta_og'],
            iris_merged['etiqueta']
            )
```

```python
iris_merged['Exactitud'].value_counts(normalize=True)
```

```python
iris_etiqueta_check = (iris_merged
                        .groupby(['etiqueta_og'])
                        .agg({'Exactitud':'sum'})
                        .merge(iris_merged['etiqueta_og']
                                .value_counts()
                                , how='left'
                                , left_index=True
                                , right_index=True
                                )
                                )
iris_etiqueta_check['Porcentaje'] = np.round(iris_etiqueta_check['Exactitud'] / iris_etiqueta_check['count'],4)
iris_etiqueta_check
```
